```python

def euclidean_dist(ptA, ptB):
    # compute and return the euclidean distance between the two points
    return np.linalg.norm(ptA - ptB)

def eye_aspect_ratio(eye) :
    A = dist.euclidean(eye[1], eye[5])
    B = dist.euclidean(eye[2], eye[4])
    C = dist.euclidean(eye[0], eye[3])
    ear = (A + B) / (2.0 * C)
    return ear

'''def init_message() :
    print("init_message")
    alarm.sound_alarm("init_sound.mp3")'''
    
import winsound

def init_message() :
    print("init_message")
#     sound_alarm("init_sound.mp3")


    freq = 2500 # Set frequency To 2500 Hertz
    dur = 1000 # Set duration To 1000 ms == 1 second

    winsound.Beep(freq, dur)
    winsound.Beep(freq, dur)
    winsound.Beep(freq, dur)
    winsound.Beep(freq, dur)
    winsound.Beep(freq, dur)
    winsound.Beep(freq, dur)
    winsound.Beep(freq, dur)
    winsound.Beep(freq, dur)
    winsound.Beep(freq, dur)
```

```python
def light_removing(frame) :
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    lab = cv2.cvtColor(frame, cv2.COLOR_BGR2LAB)
    L = lab[:,:,0]
    med_L = cv2.medianBlur(L,99)
    #median filter
    #관심픽셀 주변으로 지정한 커널 크기( 99 x 99) 내의 픽셀을 크기순으로 정렬한 후 중간값을 뽑아서 픽셀값으로 사용.

    invert_L = cv2.bitwise_not(med_L) 
    #invert lightness
    composed = cv2.addWeighted(gray, 0.75, invert_L, 0.25, 0)
    # cv2.addWeighted 함수의 인자가 (imgA, a, imgB,b, c)라고 한다면
    # result = imgA * a + imgB * b + c
    # 즉 우리 상황에서는
    # gray이미지 75% inver_L 25%
    return L, composed
```

```python
def image(frame):
    
    try:   

        faces = face_cascade.detectMultiScale(frame, 1.15, 6) 
        x,y,w,h = faces[0]
        y += 18
        h += 5
        x += 8
        w -= 10
        gray = cv2.resize(frame[y:y+h, x:x+w],dimensions)
        imgd = cv2.cvtColor(gray, cv2.COLOR_BGR2GRAY)
        imgd = imgd / 255
        imgd = imgd.reshape(-1,dimensions[0],dimensions[1],1)

        lands = model.predict(imgd)

        lands = np.round((lands.reshape(-1, 2) * 48) + 48)
                
        for land in lands:
            cv2.circle(gray, (int(land[0]), int(
                    land[1])), 1, (0, 255, 0), 1)
            
        gray =  cv2.resize(gray,(w,h))
        frame[y:y+h, x:x+w,] = gray

        return frame

    except:
        return frame
```
